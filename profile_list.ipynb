# generate a list of profile and pages using the ig resource file

1. get ig.yaml file
2. create list of profiles
   1. find all structuredefintion ids
   2. prepend the base path and append html to create the web address
3. create list of pages
   1. omit the toc, and artifacts pages since they are generated
   

```python
from pathlib import Path
from yaml import load, SafeLoader
base_path = 'https://build.fhir.org/ig/HL7/US-Core/'
source_dir = '/Users/ehaas/Documents/FHIR/US-Core/input/data/ig.yml'  # US Core
source_path = Path(source_dir)
source_path
```

```python
ig_obj = load(source_path.read_text(), Loader=SafeLoader)
canonical_url = ig_obj['url'].rsplit('/', 2)[0] + '/'
canonical_url
```

```python
profile_list = []
for i in ig_obj['definition']['resource']:
  if 'StructureDefinition' in i['reference']['reference']:
   profile_list.append(f"{base_path}{'-'.join(i['reference']['reference'].rsplit('/',1))}.html")
profile_list
```

```python
page_list = []  #skip first one which is toc.html
for i in ig_obj['definition']['page']['page']:
  # print(i)
  page_list.append(f"{base_path}{i['nameUrl']}")
  try:
    for j in i['page']:
      # print(j)
      page_list.append(f"{base_path}{j['nameUrl']}")
  except KeyError as e:
      pass
      # print(f"'FOO: ' {e}")
  else:
    try:
      for k in j['page']:
        # print(k)
        page_list.append(f"{base_path}{k['nameUrl']}")
    except KeyError as e:
      pass
      # print(f"'BAR: ' {e}")
    else:
      try:
        for m in k['page']:
          # print(m)
          page_list.append(f"{base_path}{m['nameUrl']}")
      except KeyError as e:
        pass
        # print(f"'BAZ: ' {e}")
page_list
```
